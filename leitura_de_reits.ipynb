```python
import requests
import pandas as pd

#https://stockmarketmba.com/listofreits.php

#Reits
#A url que você quer acesssar
#url_02 = "https://www.reit.com/investing/reit-directory"
url_preco_01 = "https://www.reit.com/investing/reit-directory?sector=All&status=309&country=8&page=0"
url_preco_02 = "https://www.reit.com/investing/reit-directory?sector=All&status=309&country=8&page=1"
url_preco_03 = "https://www.reit.com/investing/reit-directory?sector=All&status=309&country=8&page=4"
url_preco_04 = "https://www.reit.com/investing/reit-directory?sector=All&status=309&country=8&page=6"
url_preco_05 = "https://www.reit.com/investing/reit-directory?sector=All&status=309&country=8&page=8"
url_preco_06 = "https://www.reit.com/investing/reit-directory?sector=All&status=309&country=8&page=11"
url_01 = "https://stockmarketmba.com/listofreits.php"

#Informações para fingir ser um navegador
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}
#juntamos tudo com a requests
r_01 = requests.get(url_preco_01, headers=header)
r_02 = requests.get(url_preco_02, headers=header)
r_03 = requests.get(url_preco_03, headers=header)
r_04 = requests.get(url_preco_04, headers=header)
r_05 = requests.get(url_preco_05, headers=header)
r_06 = requests.get(url_preco_06, headers=header)

r_dy_01 = requests.get(url_01, headers=header)

#E finalmente usamos a função read_html do pandas
df_reit_preco_01 = pd.read_html(r_01.text)
df_reit_preco_02 = pd.read_html(r_02.text)
df_reit_preco_03 = pd.read_html(r_03.text)
df_reit_preco_04 = pd.read_html(r_04.text)
df_reit_preco_05 = pd.read_html(r_05.text)
df_reit_preco_06 = pd.read_html(r_06.text)

df_reit_dy_01 = pd.read_html(r_dy_01.text)

df_reit_01 = df_reit_preco_01[0]
df_reit_02 = df_reit_preco_02[0]
df_reit_03 = df_reit_preco_03[0]
df_reit_04 = df_reit_preco_04[0]
df_reit_05 = df_reit_preco_05[0]
df_reit_06 = df_reit_preco_06[0]

df_reit_dy_01 = df_reit_dy_01[0]

df_reits = pd.concat([df_reit_01, df_reit_02, df_reit_03, df_reit_04, df_reit_05, df_reit_06]) 
```
