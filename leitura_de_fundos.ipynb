```python
import requests
import pandas as pd

#A url que você quer acesssar
url = "https://www.fundsexplorer.com.br/ranking"

#Informações para fingir ser um navegador
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}
#juntamos tudo com a requests
r = requests.get(url, headers=header)
#E finalmente usamos a função read_html do pandas
dfs = pd.read_html(r.text)
```

```python
data = {'papel': ["MXRF11", "TGAR11", "VINO11", "MGFF11", "IRDM11", "PORD11", "MALL11", "RECT11", "VISC11", "XPCI11", "RECR11", "TORD11", "HCTR11", "RZTR11", 
                           "NCHB11", "VGIP11", "TRXF11", "KISU11", "DEVA11", "SDIL11", "GGRC11", "VSLH11", "VGHF11"],
                'qtd': [1850, 230, 100, 170, 135, 180, 140, 178, 145, 175, 138, 950, 150, 70,
                        36, 10, 35, 500, 22, 40, 40, 200, 70]
        }

df_meus_fiis = pd.DataFrame(data)

print (df_meus_fiis)
```

```python
df_fii = dfs[0]
```

```python
df_fii.rename(columns = {'Códigodo fundo':'papel', 
                         'Preço Atual':'preco',
                         'Liquidez Diária':'liquidez',
                         'Dividendo': 'div',
                         'DividendYield': 'dy_mes',      
                         'DY (12M)Média': 'dy_ano_media',
                         'DY Ano':  'dy_ano',
                         'PatrimônioLíq.': 'patrimonio',
                         'VPA':  'vpa',
                         'P/VPA':  'p_vpa',         
                         'Setor':  'setor',
                         'QuantidadeAtivos': 'qtd_ativos'}, inplace = True)
```

```python
columns = ['papel', 'setor', 'preco', 'liquidez', 'div', 'dy_mes', 'dy_ano_media', 'dy_ano', 'patrimonio', 'vpa', 'p_vpa', 'qtd_ativos']
df_fii_02 = pd.DataFrame(df_fii, columns=columns)
```

```python
df_fii_03 = df_fii_02.loc[df_fii_02['papel'].isin(['MXRF11', 'TGAR11', 'VINO11', 'MGFF11', 'IRDM11', 'PORD11', 'MALL11', 'RECT11', 'VISC11', 'XPCI11', 'RECR11',
                                        'TORD11', 'HCTR11', 'RZTR11', 'NCHB11', 'VGIP11', 'TRXF11', 'KISU11', 'DEVA11', 'SDIL11', 'GGRC11', 'VSLH11', 'VGHF11'])]   
```

```python
df_fii_03.head(25)
```

```python
 df_fundos = pd.merge(df_fii_03, df_meus_fiis, on=["papel"])
```

```python
df_fundos.dy_mes = df_fundos.dy_mes.str.replace('[%]', '')
df_fundos.dy_mes = df_fundos.dy_mes.str.replace(',', '.') 

df_fundos['div'] = df_fundos['div'].str.replace('[R$ ]','', regex=True)
df_fundos['div'] = df_fundos['div'].str.replace(',','.', regex=True)

df_fundos['preco'] = df_fundos['preco'].str.replace('[R$ ]','', regex=True)
df_fundos['preco'] = df_fundos['preco'].str.replace(',','.', regex=True)
```

```python
df_fundos.head()
```

```python
df_fundos['div'] = df_fundos['div'].astype(float, errors = 'raise')
df_fundos['dy_mes'] = df_fundos['dy_mes'].astype(float, errors = 'raise')
df_fundos['preco'] = df_fundos['preco'].astype(float, errors = 'raise')
```

```python
df_fundos["total_dy"] = df_fundos["div"] * df_fundos["qtd"]
df_fundos["meu_patrimonio"] = df_fundos["preco"] * df_fundos["qtd"]

df_fundos.head()
```

```python
df_fundos['posicao'] = round((df_fundos['meu_patrimonio']/df_fundos['meu_patrimonio'].sum())*100, 1)
```
